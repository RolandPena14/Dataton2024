# Datos del concurso de Admisión entre los años 2021-I  y 2024-I

En el marco de la Datatón Exprésate Perú con Datos 2024, se propone un análisis exhaustivo de los datos de los **postulantes a la Universidad Nacional de Ingeniería** (UNI) durante los procesos de admisión desde 2021-I hasta 2024-I. Este estudio tiene como objetivo identificar patrones, tendencias y áreas de mejora en el proceso de admisión, con el fin de contribuir a la solución de problemas educativos en el Perú. Además, se busca alinear las conclusiones del análisis con los Objetivos de Desarrollo Sostenible (ODS) como:

*   **ODS 4**: Garantizar una educación inclusiva, equitativa y de calidad y promover oportunidades de aprendizaje durante toda la vida para todos.
*   **ODS 5**: Lograr la igualdad entre los géneros y empoderar a todas las mujeres y las niñas.

Mediante este análisis se busca entender mejor las dinámicas del proceso de admisión en la UNI, así como también proponer soluciones concretas para fomentar la equidad de género y mejorar la calidad educativa en todo el país. Este estudio servirá como base para futuras estrategias educativas y programas de intervención, con el objetivo de que más estudiantes, independientemente de su género y origen de entidad educativa, tengan acceso a una educación de calidad y la oportunidad de ser admitidos en universidad de alta calidad educativa como la UNI.



## **Problema general**

El análisis busca responder las siguientes preguntas clave para entender mejor el perfil y desempeño de los postulantes a la UNI:


*   **Origen Geográfico de los Colegios**: ¿Cuál es la distribución geográfica de los postulantes? ¿De qué departamentos provienen la mayoría de los estudiantes?
*   **Demanda por Especialidades:** ¿Cuáles son las especialidades más solicitadas por los postulantes?
* **Participación Femenina**: ¿Cuál es el porcentaje de mujeres entre los postulantes y cómo ha evolucionado este a lo largo de los años?
* **Edad Promedio**: ¿Cuál es la edad promedio de los postulantes y qué implica esto para su formación académica y su futura inserción laboral?



## **Problemas específicos a resolver**

*	**Tasa de Admisión y Origen Geográfico de sus Colegios**: Analizar la tasa de admisión de los postulantes y su relación con el lugar de procedencia de sus colegios.
*	**Factores Asociados a la Admisión**: Identificar las características demográficas y académicas que influyen en la admisión de los postulantes.
*	**Tendencia en las Modalidades de Ingreso**: Evaluar la tendencia en las diferentes modalidades de ingreso y cómo estas se relacionan con las tasas de admisión.
*	**Rendimiento Académico y Procedencia Geográfica**: Analizar el rendimiento académico de los postulantes en función del lugar de procedencia de sus colegios.

### **2.1 Problema General**



#### 2.1.1. **Cantidad de estudiantes provenientes de colegios de los departamentos del Perú**
Vamos a mostrar la cantidad de postulantas al examen de admisión de la UNI en un mapa para ver la distribución territorial de los estudiantes en el Perú.

```python
import geopandas as gpd
from mpl_toolkits.axes_grid1 import make_axes_locatable

# Shapefile de los departamentos
mapa = gpd.read_file('/content/drive/MyDrive/Datos-UNI/Departamental_INEI_2023.shp') # Dirección del archivo en formado shapefile

# Cantidad de alumnos por deparamento de domicilio
departamentos = df['COLEGIO_DEPA'].value_counts().reset_index()
departamentos.columns = ['DEPARTAMENTO', 'CANTIDAD']

# Unir la información del mapa con la cantidad de postulantes
mapa = mapa.merge(departamentos, left_on='DEPARTAMEN', right_on='DEPARTAMENTO')

# Graficar el mapa con un heatmap
fig, (ax, ax1) = plt.subplots(1, 2, figsize=(15, 5))
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)

# Gráfico del mapa del Perú
mapa.plot(column='CANTIDAD', ax=ax, legend=True, cax=cax, edgecolor='Grey', linewidth=1,cmap='viridis', norm=matplotlib.colors.LogNorm(vmin=mapa.CANTIDAD.min(), vmax=mapa.CANTIDAD.max()), legend_kwds={'label': "Cantidad de Alumnos"})
ax.set_xlabel('Longitud', fontsize=10)
ax.set_ylabel('Latitud', fontsize=10)

mapa = mapa.sort_values(by = 'CANTIDAD', ascending = False)
barlist= ax1.bar(mapa['DEPARTAMENTO'], mapa['CANTIDAD'])
ax1.set_xticklabels(labels=mapa['DEPARTAMENTO'], rotation=90, ha='center')
# rects1 = ax.bar(x - width/2, men_means, width, label='Men')
ax1.bar_label(barlist, padding=3, rotation=90)
ax1.set_ylim(0,37000)
barlist[0].set_color('r')
barlist[5].set_color('r')
# ax1.grid()

plt.figtext(0.5, 1.0, 'Distribución Geográfica de los Postulantes por Departamento', fontsize=15, ha='center', va='center')
plt.show()
```

```python
# Calcular porcentajes de la cantidad de estudiantes provenientes de los distintos departamentos
mapa['PORCENTAJE'] = round((mapa['CANTIDAD'] / mapa['CANTIDAD'].sum()) * 100, 1)
mapa.head(10)
```

Lima Metropolitana es el departamento donde se concentran la mayor cantidad de los estudiantes del Centro Preuniversitario con el 78.4%. En segundo lugar, la Provincia Constitucional del Callao con el 4.8%. En tercer lugar tenemos a Junín con el 1.9%.

## **3. TRANSFORMACIÓN DE DATOS**
En este análisis vamos a enfocarnos en los estudiantes que vieven en el Perú, entonces en función del lugar de domicilio vamos a seleccionar a los estudiantes que viven o tienen como dirección en territorio nacional.

```python
# Evaluar departamentos donde viven los estudiantes
print("Departamentos antes de la limpieza :", df['COLEGIO_DEPA'].unique())
# Eliminar los valores que no corresponden a Perú tales como: VENEZUELA, ARGENTINA...
df = df[df["COLEGIO_DEPA"].str.contains("VENEZUELA|ARGENTINA|ECUADOR|CANADA|ESPAÑA|CHILE|ITALIA|COLOMBIA") == False]
# Corregir los nombres de los departamentos eliminando las tildes para evitar dos variables distintas
df['COLEGIO_DEPA'] = df['COLEGIO_DEPA'].map({'ÁNCASH': 'ANCASH', 'HUÁNUCO': 'HUANUCO', 'APURÍMAC': 'APURIMAC', 'JUNÍN': 'JUNIN', 'SAN MARTÍN': 'SAN MARTIN'}).fillna(df['COLEGIO_DEPA'])
print("Departamentos después de la limpieza :", df['COLEGIO_DEPA'].unique())
```

```python
# Añadir columnas calculadas para conocer la edad de los estudiantes y los años de preparación
df['ANIO_PREPA'] = df['ANIO_POSTULA'] - df['COLEGIO_ANIO_EGRESO']
df['EDAD_POSTULA'] = df['ANIO_POSTULA'] - df['ANIO_NACIMIENTO']
```

```python
# Eliminar variables que no se utilizarán
df = df.drop(columns=['COLEGIO', 'COLEGIO_PROV', 'COLEGIO_DIST', 'COLEGIO_PAIS', 'COLEGIO_ANIO_EGRESO',
                      'ANIO_POSTULA', 'CICLO_POSTULA',
                      'NACIMIENTO_DEPA', 'NACIMIENTO_PROV', 'NACIMIENTO_DIST', 'ANIO_NACIMIENTO', 'NACIMIENTO_PAIS',
                      'DOMICILIO_DEPA', 'DOMICILIO_PROV', 'DOMICILIO_DIST'])
df.head(5)
```

```python
# Evaluar valores nulos nuevamente porque pueden perjudicar la corrida de los modelos
df.isnull().sum()
```

```python
# Eliminar los valores nulos
df = df.dropna()
```

### **4.1 Problemas Específicos**



#### 4.1.1. **Conocer la tasa de admisión de los postulantes y su relación con el lugar de origen de sus colegios**



```python
# Agrupar por departamento de origen del colegio, sexo y admisión a la UNI
admission_rate = df.groupby(['COLEGIO_DEPA', 'SEXO'])['INGRESO'].value_counts(normalize=True).unstack().fillna(0)

# Gráficar tasa de admisión por género y departamento de origen del colegio
admission_rate.plot(kind='barh', stacked=True, color=['red', 'green'], figsize=(14, 10))
plt.title('Tasa de Admisión por Departamento del Colegio y Género')
plt.ylabel('Tasa de Admisión')
plt.xlabel('Departamento y Género')
plt.legend(title='Admitido', loc='upper right', bbox_to_anchor=(1.08, 1))
plt.xlim(0,1)
plt.show()
```

```python
# Ordenar en función de la tasa de admisión
admission_rate.sort_values(by = 'SI', ascending = False).head(5)
```

Los ratios de admisión por colegio de origen y sexo son:
1. TUMBES (FEMENINO): 33.3 %
2. JUNIN (MASCULINO): 21.2 %
3. LAMBAYEQUE (MASCULINO): 20.2 %

#### 4.1.2. **Averiguar una tendencia en las modalidades de ingreso asociada con la admisión en la UNI**


```python
# Análisis de la efectividad de las modalidades de ingreso
mode_success_rate = df.groupby('MODALIDAD')['INGRESO'].value_counts(normalize=True).unstack().fillna(0)

# Gráfico de la tasa de éxito por modalidad de ingreso
mode_success_rate.plot(kind='barh', stacked=True, color=['blue', 'orange'], figsize=(14, 10))
plt.title('Tasa de Éxito por Modalidad de Ingreso')
plt.ylabel('Tasa de Éxito')
plt.xlabel('Modalidad de Ingreso')
plt.legend(title='Admitido', loc='upper right', bbox_to_anchor=(1.08, 1))
plt.xlim(0,1)
plt.show()
```

```python
mode_success_rate.sort_values(by = 'SI', ascending = False).head(5)
```

#### 4.1.3. **Conocer las características asociadas a la admisión a la UNI de los postulantes**

```python
# Importar librerías de Machine Learning
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns

# Eliminar valores nulos
df = df.dropna()

# Preprocesamiento: Codificar variables categóricas
le = LabelEncoder()
df['SEXO'] = le.fit_transform(df['SEXO'])
df['MODALIDAD'] = le.fit_transform(df['MODALIDAD'])
df['INGRESO'] = le.fit_transform(df['INGRESO'])

# Seleccionar características de X y la variable objetivo y
X = df[['SEXO', 'EDAD_POSTULA', 'ANIO_PREPA', 'CALIF_FINAL', 'MODALIDAD']]
y = df['INGRESO']

# Dividir en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42) # con 30% de datos de test

# Entrenar un modelo de Random Forest
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Hacer predicciones
y_pred = model.predict(X_test)

# Evaluar el modelo
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred)) # el modelo muestra un buen rendimiento (91%)

# Importancia de las características
importances = model.feature_importances_
sns.barplot(x=importances, y=X.columns)
plt.title('Importancia de las Características')
plt.show()
```

#### 4.1.4. **Conocer el rendimiento académico según el lugar de procedencia del colegio**


```python
# Análisis del rendimiento académico por ubicación del colegio
avg_by_dep = df.groupby('COLEGIO_DEPA')['CALIF_FINAL'].mean()

# Gráfico del rendimiento académico por departamento
avg_by_dep.sort_values(ascending = False).plot(kind='bar', color='purple', figsize=(14, 6))
plt.title('Rendimiento Académico Promedio por Departamento del Colegio')
plt.ylabel('Calificación Final Promedio')
plt.xlabel('Departamento del Colegio')
plt.show()
```

```python
avg_by_dep.sort_values(ascending = False).head(5)
```

Los promedios de las calificaciones finales por departamento son:
